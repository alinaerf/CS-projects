```python
class MaxHeapq:
    """ 
    A class that implements properties and methods 
    that support a max priority queue data structure

    Attributes
      ----------
      heap : arr
          A Python list where key values in the max heap are stored
      heap_size: int
          An integer counter of the number of keys present in the max heap
      """  

    def __init__(self,heap):    
        """
        Parameters
        ----------
        None
        """    
        self.heap       = heap
        self.heap_size  = len(heap)
        
    def left(self, i):
        """
        Takes the index of the parent node
        and returns the index of the left child node

        Parameters
        ----------
        i: int
          Index of parent node

        Returns
        ----------
        int
          Index of the left child node

        """
        return 2 * i + 1

    def right(self, i):
        """
        Takes the index of the parent node
        and returns the index of the right child node
        
        Parameters
        ----------
        i: int
            Index of parent node

        Returns
        ----------
        int
            Index of the right child node

        """

        return 2 * i + 2
    
    def parent(self, i):
        """
        Takes the index of the child node
        and returns the index of the parent node
        
        Parameters
        ----------
        i: int
            Index of child node

        Returns
        ----------
        int
            Index of the parent node

        """

        return (i - 1)//2

    def maxk(self):     
        """
        Returns the highest key in the priority queue. 
        
        Parameters
        ----------
        None

        Returns
        ----------
        int
            the highest key in the priority queue

        """
        return self.heap[0]     
    
  
    def heappush(self, key):  
        """
        Insert a key into a priority queue 
        
        Parameters
        ----------
        key: int
            The key value to be inserted

        Returns
        ----------
        None
        """
        add=Task(-float("inf"),"a",0,[],-float("inf"))
        self.heap.append(add)
        self.increase_key(self.heap_size,key)
        self.heap_size+=1 
        
        
    def increase_key(self, i, key): 
        """
        Modifies the value of a key in a max priority queue
        with a higher value
        
        Parameters
        ----------
        i: int
            The index of the key to be modified
        key: int
            The new key value

        Returns
        ----------
        None
        """
       
        if key.priority < self.heap[i].priority:
            raise ValueError('new key is smaller than the current key')
        self.heap[i] = key
        while i > 0 and self.heap[self.parent(i)].priority < self.heap[i].priority:
            j = self.parent(i)
            holder = self.heap[j]
            self.heap[j] = self.heap[i]
            self.heap[i] = holder
            i = j    
            
       
    def heapify(self, i):
        """
        Creates a max heap from the index given
        
        Parameters
        ----------
        i: int
            The index of of the root node of the subtree to be heapify

        Returns
        ----------
        None
        """
        l = self.left(i)
        r = self.right(i)
        heap = self.heap
        if l <= (self.heap_size-1) and heap[l]>heap[i]:
            largest = l
        else:
            largest = i
        if r <= (self.heap_size-1) and heap[r] > heap[largest]:
            largest = r
        if largest != i:
            heap[i], heap[largest] = heap[largest], heap[i]
            self.heapify(largest)

    def heappop(self):
        """
        returns the largest key in the max priority queue
        and remove it from the max priority queue
        
        Parameters
        ----------
        None

        Returns
        ----------
        int
            the max value in the heap that is extracted
        """
        if self.heap_size < 1:
            raise ValueError('Heap underflow: There are no keys in the priority queue ')
        maxk = self.heap[0]
        self.heap[0] = self.heap[-1]
        self.heap.pop()
        self.heap_size-=1
        self.heapify(0)
        return maxk
```

A. The main changes in code will take place in the function run_task_scheduler. When the program uses the heappop  to extract the task with the highest priority, the code additionally checks whether the task can be multitasked. If so, the program needs to go through all other tasks in the current heap to check whether there are any activities that can be combined with the task. The program does not only check the top node of the heap but rather goes through all activities in the heap because all of them can be executed (they have no dependencies if they are in a heap). If the second activity is found, the loop terminates and executes both activities at the same time, taking the maximum time for two activities (e.g., if task 1 takes more time, its duration is added to the current time and vice versa). If the second activity for multitasking is not found, the program executes the first found task and continues its work. 

B. The constraint of multitasking makes the primary function prioritize multitasking over the usual MaxHeapq structure. When the first extracted task is found to be multitaskable, the program goes through all tasks in a heap instead of checking only the top node because if two tasks can be at least partially be completed together, it is more time-efficient to not focus on priority. Further, there are some additions to function heappop as it needed adaptions to be able to extract nodes other than the top one. An additional argument will be added to the function, which specifies which node to execute, but by default, it would be set to 0.

```python
"""
Cited from session 5.2. 

Link: https://sle-collaboration.minervaproject.com/?url=h
ttps%3A//sle-authoring.minervaproject.com/api/v1/worksheets/a55439fa-65fc-403a-9d1a
-f532c4d84b60/&userId=10930&name=Alina+Erofeeva&avatar=https%3A//s3.amazonaws.com/pic
asso.fixtures/Alina_Erofeeva_10930_2021-01-22T07%3A20%3A32.132Z&noPresence=1&readOnly=1
&isInstructor=0&signature=8011b77e8142e1b66ddb3a0da55e1e6c7245c157fef81eac3af7fc4d71807d80

"""
class Task:
    """
    - id: Task Id   
    - description: Short description of the task   
    - duration: Duration in minutes 
    Duration of the task is used to determine the priority of the task.
    - priority: Priority level of a task    
    - status: Current status of the task
    - type: Type of the performed activity
        "AC"- academic
        "TR"- transportation
        "ENT" - entertainment
    Task type is used to determine the priority of the task.
     - mutitasking: Boolean variable that signifies whether an activity can be multitasked 
        
    """
    
    #Initializes an instance of Task
    def __init__(self,task_id,description,duration,dependencies,task_type,multitasking=False,status="N",priority=0):
        
        self.id= task_id
        self.description=description
        self.duration=duration
        self.dependencies=dependencies
        self.status=status
        self.priority=priority
        self.type=task_type
        self.multitasking=multitasking
        
        
    def __repr__(self):
        return f"{self.description} - id: {self.id}\n \tDuration:{self.duration}\n\tDepends on: {self.dependencies} \
        \n\tTask Type: {self.type}\n\tMultitasking: {self.multitasking}\n\tStatus: {self.status}"

    def __lt__(self, other):
        return self.id < other.id    
    
class TaskScheduler:
    """
    A Daily Task Scheduler Using Priority Queues Based on the Priority Points
    """
    #Creation of three states of the task, NOT_STARTED, IN_PRIORITY_QUEUE, COMPLETED
    NOT_STARTED ='N'
    IN_PRIORITY_QUEUE = 'I'
    COMPLETED = 'C'
    
    def __init__(self, tasks):
        self.tasks = tasks
        
        #Creating an instance of MaxHeapq to use priority queues to get the order of execution
        self.aheap=[]
        self.priority_queue = MaxHeapq(self.aheap)  
        
    #a function that prints information about all given tasks in an appropriate manner    
    def print_self(self):
        print('Input List of Tasks')
        for t in self.tasks:
            print(t)            
            
    def remove_dependency(self, task_id):
        """
        Removes dependcy on the completed task from other tasks
        
        Input: list of tasks and task_id of the task just completed
        Output: lists of tasks with t_id removed
        """
        #go through all tasks to check if there are any tasks dependent on the given task_id
        for t in self.tasks:
            #check whether it is not the same task and whether the checked task has a dependency on task_id
            if t.id != task_id and task_id in t.dependencies:
                #remove the satisfied dependency
                t.dependencies.remove(task_id)           
            
    def get_tasks_ready(self):
        """ 
        Implements step 1 of the scheduler
        Input: list of tasks
        Output: list of tasks that are ready to execute 
        (i.e. tasks with no pendending task dependencies)
        """
        #goes through all tasks
        for task in self.tasks:
            
            # If task has no dependencies and is not yet in queue
            if task.status == self.NOT_STARTED and not task.dependencies: # If task has no dependencies and is not in queue
                # Change status of the task
                task.status = self.IN_PRIORITY_QUEUE # Change status of the task
                self.priority_assignment(task)
                # Push task into the priority queue 
                self.priority_queue.heappush(task)
                
    
    def check_unscheduled_tasks(self):
        """
        Checks whether there are any unstarted tasks
        Input: list of tasks 
        Output: boolean 
        (checks the status of all tasks and returns True if at least one task has status = 'N'
        """
        for task in tasks:
            if task.status == self.NOT_STARTED:
                return True
        return False   
    
    #time formatting function for convinient time display 
    def format_time(self, time):
        #check if the tasks go beyond 24h format
        if time//60>=24:
            return f"{(time-1440)//60}h{(time-1440)%60:02d}"
        return f"{time//60}h{time%60:02d}"
    
    
    def priority_assignment(self,task):
        """
        Assigns priority to tasks based on their type and number of dependencies
        Input: task 
        Output: None
        """
        #if the task is academic, give it priority of 40 points
        if task.type=="AC":
            task.priority+=40
        #if the task is entertainment or transportation, give it priority of 25 points
        if task.type=="ENT" or task.type=="TR":
            task.priority+=25
            
        #assign additional priority based on the duration of the task 
        #attempting to finish longer tasks first
        task.priority+=task.duration*1.5
        #assign additional priority based on the number of tasks dependent on it
        for t in self.tasks:
                if t.id != task.id and task.id in t.dependencies:
                    task.priority+=5
    
    def run_task_scheduler(self, starting_time = 480):
        """
        Runs the task scheduler and prints the order of tasks/time of start and finish of the task
        Input: starting time (pre-assigned to 8am=480 minutes) 
        Output: None
        """
        #keep track of the current time by setting current time to the beginning of the day
        current_time = starting_time
        
        #check whether there are any uncompleted tasks or any tasks in the priority queue
        while self.check_unscheduled_tasks() or self.priority_queue.heap_size!=0:
            #Extract tasks ready to execute (those without dependencies) and push them into the priority queue
            self.get_tasks_ready()
            
            #check whether the day has ended, and if so, start a new day
            if current_time>1440:
                #count the hours it took to complete tasks in the first day
                total_time = current_time - starting_time
                print(f"‚è∞Simple Scheduler at time {self.format_time(current_time)} started executing task Sleep")
                current_time=480            
                print(f"‚úÖ Completed Task Sleep at time {self.format_time(current_time)}\n") 
            if self.priority_queue.heap_size > 0 :  # Check for tasks in the priority queue.      
                # get the tasks on top of the priority queue (1 line of code required)   
                task = self.priority_queue.heappop()
                
                #check if the given task can be multitasked
                if task.multitasking==True:
                    
                    #go through the heap to see if there are any tasks that can be multitasked
                    for i in range(self.priority_queue.heap_size):
                        
                        #if such task is found
                        if self.priority_queue.heap[i].multitasking==True:
                            
                            #perform heappop on the second multitasked task
                            task1 = self.priority_queue.heappop(i)
                            
                            #find the longest task between the two multitasked ones
                            if task.duration>=task1.duration:
                                overall_length=task.duration
                            else: 
                                overall_length=task1.duration
                                
                            #print the information about the two multitasked ones
                           
                            print(f"‚è∞Simple Scheduler at time {self.format_time(current_time)} started executing task {task.id} \
                            and {task1.id} that takes {overall_length} mins")
                            current_time += overall_length           
                            print(f"‚úÖ Completed Task {task.id} - '{task.description}' and {task1.id} - '{task1.description}' \
                            at time {self.format_time(current_time)}\n") 
                            
                            #remove the dependencies on two performed tasks and change their status to 'completed' 
                            self.remove_dependency(task1.id)
                            task1.status = self.COMPLETED
                            self.remove_dependency(task.id)
                            task.status = self.COMPLETED
                            
                            #break the for loop when the second task for multitasking is found
                            break
                    #if the second task for multitasking is not found, perform only one task
                    if not(task.status == self.COMPLETED):
                        #print information about the completed task
                        print(f"‚è∞Simple Scheduler at time {self.format_time(current_time)} started executing task {task.id} \
                        that takes {task.duration} mins")
                        current_time += task.duration            
                        print(f"‚úÖ Completed Task {task.id} - '{task.description}' at time {self.format_time(current_time)}\n") 
                
                #if the first task cannot be multitasked, perform it anyway
                else:
                    #print information about the completed task
                    print(f"‚è∞Simple Scheduler at time {self.format_time(current_time)} started executing task {task.id} \
                    that takes {task.duration} mins")
                    current_time += task.duration            
                    print(f"‚úÖ Completed Task {task.id} - '{task.description}' at time {self.format_time(current_time)}\n") 
                
                # if the task is completed, it cannot be a dependency on other tasks, so remove it from the dependency list 
                self.remove_dependency(task.id)
                task.status = self.COMPLETED
                
        #print how long it took to complete all tasks
        total_time += (current_time - starting_time)             
        print(f"üèÅ Completed all planned tasks in {total_time//60}h{total_time%60:02d}min")



```
